```python
import pickle
import pandas as pd
import numpy as np
```

```python
# Read in pmdarima results
with open("../data/pmdarima_cv_res_oct24.pkl", "rb") as f:
    arima = pickle.load(f)
arima = arima.sort_values(by=['zipcode', 'model'])
```

```python
arima.model.replace('Benchmark', 'Univariate', inplace=True)
arima.model.replace('Holdout Test Benchmark', 'Test - Univariate', inplace=True)
arima.model.replace('+ MFR OCC', 'MFR Occupancy', inplace=True)
arima.model.replace('Holdout Test + MFR OCC', 'Test - MFR Occupancy', inplace=True)
arima.model.replace('+ MFR OCC & Rent', 'MFR Occ & Rent', inplace=True)
arima.model.replace('Holdout Test + MFR OCC & Rent', 'Test - MFR Occ & Rent', inplace=True)
arima.model.replace('+ MFR Rent', 'MFR Rent', inplace=True)
arima.model.replace('Holdout Test + MFR Rent', 'Test - MFR Rent', inplace=True)
```

```python
arima['Market'] = arima['zipcode'].apply(lambda x: 'Atlanta' if x.startswith('3') else 'Cleveland')
arima.rename(columns={'mmape':'mape'}, inplace=True)
```

```python
models = ['Univariate', 'MFR Rent', 'MFR Occupancy', 'MFR Occ & Rent']
arima = arima[arima['model'].isin(models)]
```

```python
arima.reset_index(drop=True, inplace=True)
arima.head()
```

```python
arima_results = arima.groupby(['Market', 'model'])[['mape']].agg(['mean'])
arima_results
```
